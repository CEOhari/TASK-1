### Load Dataset and Display First Few Rows

```python
import pandas as pd

data = pd.read_csv("student_marks.csv")
print(data.head())
```

### Check Basic Information About Dataset

```python
print(data.info())
```

### Find and Handle Missing Values

```python
print(data.isnull())
```

```python
print(data.isnull().sum())
```

### Handle Missing Values (Fill with Average)

```python
data["Marks"].fillna(data["Marks"].mean(), inplace=True)
```

### Calculate Simple Statistics

```python
average_marks = data["Marks"].mean()
print("Average Marks:", average_marks)
```

### Highest Value (Maximum)

```python
highest_marks = data["Marks"].max()
print("Highest Marks:", highest_marks)
```

### Lowest Value (Minimum)

```python
lowest_marks = data["Marks"].min()
print("Lowest Marks:", lowest_marks)
```

* Average Marks: 85.0
* Highest Marks: 95
* Lowest Marks: 70
