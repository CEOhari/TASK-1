### Theory Concepts

### Why Visualize Data?
   Understanding Charts and Graphs
   Data visualization is the graphical representation of data to:

    Understand patterns quickly

    Identify trends, outliers, and relationships

    Communicate insights clearly to stakeholders

### Why visualization matters:

    Humans process visuals faster than tables

    Trends and comparisons are easier to spot

    Helps in decision-making

### Matplotlib: Creating Basic Charts in Python
    Matplotlib is the most widely used Python library for data visualization.
    
    Simple and flexible

    Works well with NumPy and Pandas

    Supports line, bar, pie, histogram, scatter plots, etc.

### Basic workflow:

Import Matplotlib

Prepare data

Choose chart type

Add labels and title

Display the plot

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.plot(x, y)
plt.title("Simple Line Chart")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

### Bar Charts

Comparing Categories (Products, Months, Cities)

Bar charts are used when:

Comparing discrete categories

Showing differences in quantities

```python
products = ["A", "B", "C"]
sales = [200, 150, 300]

plt.bar(products, sales)
plt.title("Product Sales Comparison")
plt.xlabel("Products")
plt.ylabel("Sales")
plt.show()
```

### Line Charts
Line charts show how data changes continuously over time.

Stock prices

Monthly sales trends

Website traffic over time

```python
months = ["Jan", "Feb", "Mar", "Apr"]
revenue = [100, 120, 140, 160]

plt.plot(months, revenue)
plt.title("Revenue Trend")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.show()
```

### Pie Charts
Pie charts represent proportions or percentages.

Market share

Budget allocation

Category contribution

```python
market_share = [40, 30, 20, 10]
companies = ["A", "B", "C", "D"]

plt.pie(market_share, labels=companies, autopct='%1.1f%%')
plt.title("Market Share Distribution")
plt.show()
```

### Project: Sales Data Analysis

Objective

Analyze a sales dataset to:

Calculate total sales (revenue)

Identify the best-selling product

Generate a clean, formatted report with insights

### Day-Wise Implementation

Day 1: Setup & Load Data

```python
import pandas as pd

# Load sales data
df = pd.read_csv("sales_data.csv")

print("Data Loaded Successfully")
print(df)
```

Day 2: Explore Data

```python
# Check number of rows and columns
print("Shape:", df.shape)

# Column names
print("Columns:", df.columns)

# Data types
print("\nData Types:")
print(df.dtypes)

# Check missing values
print("\nMissing Values:")
print(df.isnull().sum())
```

### Clean Data
Handle Missing Values & Duplicates

```python
# Fill missing Quantity with 0
df["Quantity"].fillna(0, inplace=True)

# Fill missing Price with 0
df["Price"].fillna(0, inplace=True)

# Remove duplicate rows
df.drop_duplicates(inplace=True)

print("\nCleaned Data:")
print(df)
```

### Day 4: Analyze Sales

```python
## Create Revenue Column
df["Revenue"] = df["Quantity"] * df["Price"]
```

```python
# 1. Total Revenue
total_revenue = df["Revenue"].sum()

# 2. Total Quantity Sold
total_quantity = df["Quantity"].sum()

# 3. Best-Selling Product (by revenue)
best_product = (
    df.groupby("Product")["Revenue"]
    .sum()
    .idxmax()
)

# 4. Revenue by Product
revenue_by_product = df.groupby("Product")["Revenue"].sum()

print("Total Revenue:", total_revenue)
print("Total Quantity Sold:", total_quantity)
print("Best Selling Product:", best_product)
print("\nRevenue by Product:")
print(revenue_by_product)
```

### Day 5: Create a Clean Report

```python
print("\n SALES ANALYSIS REPORT")
print("-" * 30)

print(f"Total Revenue Generated: {total_revenue:,.2f}")
print(f"Total Units Sold: {int(total_quantity)}")
print(f"Best-Selling Product: {best_product}")

print("\nRevenue Breakdown by Product:")
for product, revenue in revenue_by_product.items():
    print(f"- {product}: {revenue:,.2f}")

print("\n Key Insights:")
print(" Laptops generate the highest revenue.")
print(" Accessories like Mouse and Keyboard have steady demand.")
print(" Data quality improved after handling missing values.")
```
